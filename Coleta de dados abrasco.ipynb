```python
# 3° Passo: Já com as informações de número de páginas por eixo, construir todos os links de todas as páginas de anais daquele congresso
## As informações desses links que serão criados, serão salvas em um novo excel (também no formato json)
import pandas as pd

df2 = pd.read_excel("eixos_CBCSHS_2019.xlsx")

# Criar DataFrame vazio com colunas
df3 = pd.DataFrame(columns=["nome_eixo", "numero_eixo", "link_eixo", "link_pagina", "numero_pagina"])

# Loop sobre cada linha de df2
for index, row in df2.iterrows():
    # Loop sobre o número de páginas, começando de 0
    for i in range(0, row["paginas"]):
        # Criar a nova linha
        nova_linha = pd.DataFrame({
            'nome_eixo': [row["nome_eixo"]],
            'numero_eixo': [row["index"]],
            'link_eixo': [row["link_eixo"]],
            'link_pagina': [f'{row["link_eixo"]}&p={i}'],
            'numero_pagina': [i+1]
        })
        
        # Concatenar a nova linha ao DataFrame df3
        df3 = pd.concat([df3, nova_linha], ignore_index=True)

# Salva o dataframe em excel e json
df3.to_excel("links_trabalhos_CBCSHS_2019.xlsx", index=False)
df3.to_json("links_trabalhos_CBCSHS_2019.json", index=False)

```

```python
# 6° Passo: Unificar as informações completas em um dataframe

import pandas as pd

def unificar(nome_ano_congresso):
    resumos = pd.read_json(f"links_e_resumos_{nome_ano_congresso}.json")
    trabalhos = pd.read_json(f"trabalhos_sem_resumo_{nome_ano_congresso}.json")
    trabalhos = trabalhos.drop(columns={"resumo"}, axis=1) if "resumo" in trabalhos else 0
    unificado = pd.merge(trabalhos, resumos, on="link_resumo", how="left")
    unificado.to_excel(f"{nome_ano_congresso}_completo.xlsx", index=False)
    unificado.to_json(f"{nome_ano_congresso}_completo.json", index=False)
    unificado.to_csv(f"{nome_ano_congresso}_completo.csv", index=False)

# Raspagem de dados estará completa no arquivo CBCSHS_2019_completo.xlsx
unificar("CBCSHS_2019")
```
